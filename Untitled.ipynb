```python
class Solution:
    def generateParenthesis(self, n: int):
        if n == 0:
            return []
        result = []
        def recursion(l, r, current, stack):
            if r < l:
                return 
            if l == r ==0:
                stack.append(current)
            if l > 0:
                recursion(l-1, r, current + "(", stack)
            if r > 0:
                recursion(l, r-1, current + ')', stack)
        recursion(n, n, '', result)
        return result
Solution().generateParenthesis(3)
```
