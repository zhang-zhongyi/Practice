```python
def closest_to_zero(ints):
    # Your code goes here
    if not ints:
        return 0
    l = [abs(i) for i in ints]
    min_index = l.index(min(l))
    if ints[min_index] * -1 in ints and ints[min_index] < 0:
        return ints[min_index] * -1
    else:
        return ints[min_index]
```

```python
closest_to_zero([3,8,-8,9,-2, -6, -12, 17, 2,4])
```
