```python
class Twitter(object):

    def __init__(self):
        self.timer = itertools.count(step=-1)
        self.tweets = collections.defaultdict(collections.deque)
        self.followees = collections.defaultdict(set)

    def postTweet(self, userId, tweetId):
        self.tweets[userId].appendleft((next(self.timer), tweetId))

    def getNewsFeed(self, userId):
        tweets = heapq.merge(*(self.tweets[u] for u in self.followees[userId] | {userId}))
        return [t for _, t in itertools.islice(tweets, 10)]

    def follow(self, followerId, followeeId):
        self.followees[followerId].add(followeeId)

    def unfollow(self, followerId, followeeId):
        self.followees[followerId].discard(followeeId)
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys



# Complete the 'solve' function below.

# def solve():
#     # Write your code here
#     print("helloworld")
# if __name__ == '__main__':
#     solve()

# By reading the input format, I would design to build a pipeline with txt file to collect all the buy and sell orders received and stored into the txt file. The format will just be the exactly same as the format from the problem description with first line as the number of orders followed by each order, each at a line. My program will be able to process the txt file and output result as described. Tested successfully in my local with same input format in the txt file '{input_file_name}' __line 22.
class Orders:
    def __init__(self):
        self.file = open('{input_file_name}', 'r') # input file name with the format in the problem description that first line is number of transactions followed by each order
        self.content = self.file.readlines()
        self.counter = self.content[0][-len(self.content[0]):-1:1] 
        self.all_transactions = []
        for i in range(1, len(self.content)-1):
            self.all_transactions.append(str(self.content[i])[-len(self.content[i]):-1:1])
        
        self.all_transactions.append(self.content[-1])
        # The last line in a "txt" file will be different since there will be no space at the end of the last line, so I need to process separately.
        self.all_input_transactions = []
        
        for i in self.all_transactions:
            self.all_input_transactions.append(i.split(' '))
            
        self.d = d = {}
        for i,j in enumerate(self.all_input_transactions):
            self.d.update({i:j})
        # Convert all the orders into a hashtable/dictionary in timestamp order earliest to latest  
        # Now all the transaction data are stored in hashtable format and are ready to use.
        
        self.d2 = d2 = {}
        
    def transactions_processed(self):
        import pandas as pd
        for i in self.d.values():
            i[0] = pd.Timestamp(i[0]) # make timestamp as timestamp datatype for sorting purpose
            i[1] = int(i[1]) # convert client_id to integer data type for output
            i[3] = int(i[3]) # convert quantity size to integer data type for output
            i[5] = float(i[5]) # convert price to float data type for output
            
        self.d2 = {i:j for i, j in sorted(self.d.items(), key = lambda item: (item[1][0], item[1][5], item[1][3]))}
        # sort by time earliest to most recent, price from cheapest to highest, quantity from lowest to highest
        self.selling_orders = {k:v for k,v in self.d2.items() if v[2] == 's'}
        self.buying_orders = {k:v for k,v in self.d2.items() if v[2] == 'b'}
        
            
        from collections import deque, Counter
        self.dq = deque()
        self.ans = []
        for i in self.d2.values():
            if not self.dq:
                self.dq.append(i)
                
            unique_orders = []
            for x in self.dq:
                unique_orders.append(x[2])
            # This is to check if all the orders in the deque are same type of orders or not.    
            if len(Counter(unique_orders).keys()) < 2 and i[2] != self.dq[-1][2] and i[2] == 'b' and i[-2] != 'm': # not a market buy order
                # this means currently all the orders in the queue are selling orders and a buying order comes in
                sorteddq = sorted(self.dq, key = lambda item: item[-1]) # sort all the existed selling orders by price from low to high
                self.dq.append(i) # append the new coming order into deque first
                for j in sorteddq: # in this case j stands for selling orders
                    if i[5] >= j[5]: # new coming buy order always loop existed sell orders from the lowest price and pick the lowest prices to complete order transactions first
                    # next step is to compare quantity
                        if i[3] < j[3]: # if first selling order has fulfilled the new coming buying order 
                            j[3] -= i[3]
                            self.ans.append([i[0],i[1],j[1],j[-1],i[3]]) # output format as timestamp, buy_client_id, sell_client_id, price, size
                            self.dq.pop() # remove the buying order from original deque.
                            break # break in nested loop only stops the current inner loop, so no need to worry about the outer loop
                        elif i[3] == j[3]:
                            self.ans.append([i[0],i[1],j[1],j[-1],i[3]]) # do the same thing as above
                            self.dq.pop() # remove the buying order from original deque.
                            self.dq = deque(sorted(self.dq, key = lambda item: item[-1], reverse = True)) # make a copy of the sorteddq for dq; convert datatype back to deque after sort
                            self.dq.popleft() # since all the selling order has already been sorted price low to high
                                         # always fulfill this buying orders from left to right. If first selling order higher than this buying order
                                         # the rest selling order will also not fulfill this current buying order.
                            break # break in nested loop only stops the current inner loop, so no need to worry about the outer loop
                        else: 
                            self.dq = deque(sorted(self.dq, key = lambda item: item[-1], reverse = True)) #convert datatype back to deque after sort
                            self.dq.popleft() # remove the selling order that has been transacted from the deque
                            i[3] -= j[3] # deducted the part that the buying order has already been fulfilled and continue the loop
            
                
            elif len(Counter(unique_orders).keys()) < 2 and i[2] != self.dq[-1][2] and i[2] == 's': 
                # this means currently all the orders in the queue are buying orders and a selling order come in
                sorteddq = sorted(self.dq, key = lambda item: item[-1], reverse = True) # sort all the buying orders by price from high to low
                self.dq.append(i) # append the new coming order into deque first
                for j in sorteddq: # in this case, j stands for buying orders
                    if i[5] <= j[5]: # new sell order always loop existed buy orders from the highest price and pick the highest buy prices to complete order transactions first
                    # next step is to compare quantity
                        if i[3] < j[3]: # if first buying order has fulfilled the selling order 
                            j[3] -= i[3]
                            self.ans.append([j[0],j[1],i[1],j[-1],j[3]]) # output format as timestamp, buy_client_id, sell_client_id, price, size
                            self.dq.pop() # remove the buying order from original deque.
                            break # break in nested loop only stops the current inner loop, so no need to worry about the outer loop
                        elif i[3] == j[3]:
                            self.ans.append([j[0],j[1],i[1],j[-1],j[3]]) # do the same thing as above
                            self.dq.pop() # remove the selling order from original deque.
                            self.dq = deque(sorted(self.dq, key = lambda item: item[-1], reverse = True)) # make a copy of the sorteddq for dq; convert datatype back to deque after sort
                            self.dq.popleft() # since all the buying order has already been sorted price high to low
                                         # always fulfill selling orders from left to right. If first buying order lower than this selling order
                                         # the rest buying order will also not fulfill this current selling order.
                            break # break in nested loop only stops the current inner loop, so no need to worry about the outer loop
                        else: 
                            self.dq = deque(sorted(self.dq, key = lambda item: item[-1], reverse = True)) #convert datatype back to deque after sort
                            self.dq.popleft() # remove the buying order that has been transacted from the deque
                            j[3] -= i[3] # deducted the part that the selling order has already been fulfilled and continue the loop
        
            # Next step is to handle the queue with both unfilled buying and unfilled selling orders stucked in the queue
            elif len(Counter(unique_orders).keys()) == 2 and i[2] == 'b' and i[-2] != 'm': # a new limit buying order come in
                existed_selling_orders = [i for i in self.dq if i[2] == 's']
                existed_selling_orders = sorted(existed_selling_orders, key = lambda item: item[-1]) 
                # sort existing selling order's limit price from low to high since buy orders will always choose lowest price first.
                if i[-1] < existed_selling_orders[0][-1]: # if buy limit price still lower than lowest selling price, append the order and continue
                    self.dq.append(i)
                    continue
                else:
                    for j in existed_selling_orders: # in this case j stands for selling orders
                        if i[5] >= j[5]: # buy order always loop from the lowest price, which is the first since it is already sorted
                        # next step is to compare quantity
                            if i[3] < j[3]: # if first selling order has fulfilled the new coming buying order 
                                j[3] -= i[3]
                                self.ans.append([i[0],i[1],j[1],j[-1],i[3]]) # output format as timestamp, buy_client_id, sell_client_id, price, size
                                break
                            elif i[3] == j[3]:
                                self.ans.append([i[0],i[1],j[1],j[-1],i[3]]) # do the same thing as above
                                self.dq = deque(sorted(self.dq, key = lambda item: (item[2], -item[-1]), reverse = True)) #convert datatype back to deque after sort
                                # sorted existed orders by selling order first since letter "s" is after letter "b" in alphabetic order, I need reverse = True, price lowest first,
                                # so add a hyphen -item[-1] to keep the ascending order.
                                self.dq.popleft() # pop the first selling order in the deque that fulfilled the new coming buying order
                                break # break in nested loop only stops the current inner loop, so no need to worry about the outer loop
                            else: 
                                self.dq = deque(sorted(self.dq, key = lambda item: (item[2], -item[-1]), reverse = True)) #convert datatype back to deque after sort
                                self.dq.popleft() # remove the selling order that has been transacted from the deque
                                i[3] -= j[3] # deducted the part that the buying order has already been fulfilled and continue the loop
                    if i[3] > 0: # some of the new buy orders haven't been fulfilled (only partial filled) and no more selling order can fulfill
                        self.dq.append([i[0],i[1],i[2],i[3],i[4],i[5]]) #append the rest quantity into the deque
            
            elif len(Counter(unique_orders).keys()) == 2 and i[2] == 's': # a new selling order come in
                existed_buying_orders = [i for i in self.dq if i[2] == 'b']
                existed_buying_orders = sorted(existed_buying_orders, key = lambda item: item[-1], reverse = True) 
                # sort existing buying order's limit price from high to low since sell orders will always choose highest price first.
                if i[-1] > existed_buying_orders[0][-1]: # if sell limit price still high than highest buying limit price, append the order and continue
                    self.dq.append(i)
                    continue
                else:
                    for j in existed_buying_orders: # in this case j stands for buying orders
                        if i[5] <= j[5]: # sell order always loop from the highest price, which is the first since it is already sorted
                        # next step is to compare quantity
                            if i[3] < j[3]: # if first buying order has fulfilled the new coming selling order 
                                j[3] -= i[3]
                                self.ans.append([i[0],i[1],j[1],j[-1],i[3]]) # output format as timestamp, buy_client_id, sell_client_id, price, size
                                break
                            elif i[3] == j[3]:
                                self.ans.append([i[0],i[1],j[1],j[-1],i[3]]) # do the same thing as above
                                self.dq = deque(sorted(self.dq, key = lambda item: (item[2], -item[-1]))) # convert datatype back to deque after sort
                                # sorted existed orders by buying order first since letter "b" is ahead of letter "s" in alphabetic order, and price highest first,
                                # so add a hyphen -item[-1] to keep the descending order.
                                self.dq.popleft() # pop the first buying order that fulfilled the new coming selling order
                                break # break in nested loop only stops the current inner loop, so no need to worry about the outer loop
                            else: 
                                self.dq = deque(sorted(self.dq, key = lambda item: (item[2], -item[-1]))) #convert datatype back to deque after sort
                                self.dq.popleft() # remove the buying order that has been transacted from the deque
                                i[3] -= j[3] # deducted the part that the selling order has already been fulfilled and continue the loop
                    if i[3] > 0: # some of the new selling orders haven't been fulfilled (only partial filled) and no more buying order can fulfill
                        self.dq.append([i[0],i[1],i[2],i[3],i[4],i[5]]) #append the rest quantity into the deque
            
            # if a market buy order come in, it will directly pick the lowest selling price orders to complete the transaction
            elif i[2] == 'b' and i[-1] < 0: # market order will be marked as -1.00 as the price
                if not [i for i in self.dq if i[2] == 's']: # if currently no selling order in the queue, append the market buy order
                    self.dq.append(i)
                else:
                    existed_selling_orders = sorted([i for i in self.dq if i[2] == 's'], key = lambda item: item[-1])
                    # sort all the existed selling order price from lowest to highest
                    # market buy order will always pick the lowest price selling order to transact.
                    for j in existed_selling_orders: # in this case j stands for selling orders
                        if i[3] < j[3]: # if first selling order has fulfilled the new coming buying order 
                            j[3] -= i[3]
                            self.ans.append([i[0],i[1],j[1],j[-1],i[3]]) # output format as timestamp, buy_client_id, sell_client_id, price, size
                            break
                        elif i[3] == j[3]: # if the selling order and buying order has same quantity
                            self.ans.append([i[0],i[1],j[1],j[-1],i[3]]) # do the same thing as above
                            self.dq = deque(sorted(self.dq, key = lambda item: (item[2], -item[-1]), reverse = True)) # convert datatype back to deque after sort
                            # sorted existed orders by selling order first since letter "s" is after letter "b" in alphabetic order, I need reverse = True, price lowest first,
                            # so add a hyphen -item[-1] to keep the ascending order.
                            self.dq.popleft() # pop the first selling order in the deque that fulfilled the new coming buying order
                            break # break in nested loop only stops the current inner loop, so no need to worry about the outer loop
                        else: 
                            self.dq = deque(sorted(self.dq, key = lambda item: (item[2], -item[-1]), reverse = True)) # convert datatype back to deque after sort
                            self.dq.popleft()
                            i[3] -= j[3] # deducted the part that the buying order has already been fulfilled and continue the loop
                    if i[3] > 0: # some of the new market buy orders haven't been fulfilled (only partial filled) and no more selling order can fulfill
                        self.dq.append([i[0],i[1],i[2],i[3],i[4],i[5]]) #append the rest quantity into the deque

        return self.ans

Orders().transactions_processed()
            
            
            
            
```
